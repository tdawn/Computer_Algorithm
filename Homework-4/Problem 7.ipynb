```python
import numpy as np
import time

def find_next_station(station_array, last_station, gas):
    i = 0
    j = 0
    temp_station = []
    km_last_station = last_station[0] 
    if km_last_station == 99999:
        dif_gas = station_array[i,0]
    else:
        dif_gas = station_array[i,0] - last_station[0]
    while dif_gas <= gas and dif_gas >= 0:
        if j == 0:
            next_station = station_array[i]
            index_next_station = j
        if dif_gas <= gas:
            temp_station.append(station_array[i,:].tolist())
            j += 1
        if temp_station[j-1][1] < next_station[1]:
            next_station = temp_station[j-1]
            index_next_station = j-1
        i += 1
        dif_gas = station_array[i,0] - km_last_station
    station_array = np.delete(station_array, range(j),axis = 0)
    if len(temp_station) > 0:
        return(station_array, temp_station, next_station)
    else:
        return("Impossible")

def lowestCost(lkm, station, station_array):
    gas = 100
    rest_gas = 100
    max_gas = 200

    last_station = [99999,99999]
    station_array, temp_station, last_station = find_next_station(station_array, last_station, gas)
    fill_gas = 0
    dif_km = last_station[0]
    dif_cost = last_station[1]
    cost = 0

    while True:
        if lkm - last_station[0] == 0:
            gas = gas - dif_km
            next_station = station_array[len(station_array)-1]
            if gas == 100:
                fill_gas = 0
            elif gas < 100:
                fill_gas = 100 - gas
            cost += next_station[1] * fill_gas
            break
        elif lkm - last_station[0] <= 200:
            gas = gas - dif_km
            next_station = station_array[len(station_array)-1]
            dif_cost = next_station[1] - last_station[1]
            dif_km = next_station[0] - last_station[0]
            if dif_km + gas <= 100 - gas:
                fill_gas = dif_km + 100
            else:
                fill_gas = dif_km
            gas += fill_gas
            cost += last_station[1] * fill_gas
            last_station = next_station[:]
        else:
            gas = gas - dif_km
            station_array, temp_station, next_station = find_next_station(station_array, last_station, 200)
            dif_cost = next_station[1] - last_station[1]
            dif_km = next_station[0] - last_station[0]
            if dif_cost <= 0:
                fill_gas = dif_km
            else:
                fill_gas = max_gas - gas
            gas += fill_gas
            cost = cost + last_station[1] * fill_gas
            last_station = next_station[:]

    return(cost)

def main():
    lkm = int(input("Enter the distance between two cities : "))
    station = int(input("Enter the number of fuel stations : "))
    print("Please input the information of the station (location and cost)")
    data = [] 
    for i in range(0, station): 
        x = input("") 
        data.append(x) 
        b = data[i].split(" ") 
        data[i] = [int(a) for a in b] 
    start = time.time()
    result = lowestCost(lkm, station, np.array(data))
    end = time.time()
    print("\n" + "===================================")
    print("\n" + "Output : " + "\n" + str(result))
    print("Time : " + str(round(end-start,5)) + " sec")
```
